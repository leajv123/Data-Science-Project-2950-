### The History of Stop-and-Frisk in NYC:

The New York Police Department’s <a href = 'https://www1.nyc.gov/site/ccrb/investigations/stop-question-and-frisk.page'>Stop, Question, and Frisk program</a> is a crime-prevention strategy in which officers temporarily detain, question, and often search individuals suspected of carrying weapons or contraband. The program gained traction in the early 2000s under Mayor Michael Bloomberg and [reached its height in 2011](https://www.nyclu.org/en/stop-and-frisk-data), a year in which an estimated 685,724 civilians were stopped by the NYPD. In August of 2013, US District Court Judge Sheila Scheindlin ruled in Floyd v. City of New York that Stop-and-Frisk had been implemented by the NYPD in an unconstitutional manner. This decision marked a turning point in the program through the subsequent NYPD mandate that all officers thoroughly justify and document the reason for each stop. Instances of Stop-and-Frisk declined in the following years due to accusations of racial profiling, as many believed the program disproportionately targeted marginalized communities, specifically African-American and Latino individuals. While the percentage of NYPD resources that was devoted to the Stop-and-Frisk program is not publicly available, the funding and manpower that must have been required to make hundreds of thousands of stops for over a decade begs the question: was the NYPD’s Stop-and-Frisk program truly effective at reducing crime in New York? Or, was it simply an act of racial discrimination designed to incite fear in the streets? 

### Research Question(s):

We are interested in evaluating crime in New York City over a 10 year period. How has crime changed in New York City from 2009 to 2019? How has the prevalence of certain crimes transformed over this timespan?

We are also interested in how implementation of certain policing programs has potentially influenced the crime statistics we find. In particular, we are interested in how the Stop, Question, and Frisk program influenced crime statistics during this period.


### Main Findings:

Our research demonstrates that the New York Police Department's Stop, Question, and Frisk program *may* have played a role in the reduction of drug- and weapon-related crimes in NYC from the height of the program in 2011 to 2019. However, our analysis also confirmed public suspicion: the NYPD's Stop and Frisk program did in fact exhibit racial profiling, targeting Black males living in the Bronx in particular.

### Linear Regression Analysis of Stop-and Frisk's Efficacy

#### Using a Linear Regression Model to Predict the Number of Annual Offenses

This linear regression model will predict the total number of criminal offenses reported annually in NYC from the number of stops recorded annually in NYC. 

The purpose of this model is to investigate whether the NYPD's Stop and Frisk Program had a tangible and quantifiable effect on criminal activity reported in NYC from 2011 to 2019.

#### Gathering the Data:

```python
# Get the total number of criminal offenses recorded each year (2011-2019)
offense_types = nyc_crime['offense'].count()

annual_offenses = []
for year in nyc_crime.columns[3:12]:
    count = 0
    for offense in range(offense_types):
        count = count + nyc_crime[year][offense]
        
    annual_offenses = annual_offenses + [count]
print('Total Criminal Offenses Reported Annually (2011-2019): ')
print(annual_offenses)
```

```python
# Get the total number of stops reported - Dataset #4

# Source: https://www.nyclu.org/en/stop-and-frisk-data

annual_stops = np.array([685724, 532911, 191851, 45787, 22565, 12404, 11629, 11008, 13450])

print('Total Stops Reported Annually (2011-2019): ')
print(annual_stops)
```

#### Performing the Linear Regression to Predict Total Offenses:

```python
# Perform linear regression with number of stops each year as predictor var X and total number of offenses as criterion var y

linear_model = LinearRegression().fit(annual_stops.reshape(-1,1), annual_offenses)

# Scatterplot of observed values and predicted values
plt.scatter(annual_stops, annual_offenses, label = 'Observed Values')
predicted_offenses = linear_model.predict(annual_stops.reshape(-1,1))
plt.scatter(annual_stops, predicted_offenses, label = 'Predicted Values', color= orange)
plt.title('Annual Stops vs. Annual Offenses')
plt.xlabel('Annual Stops')
plt.ylabel('Annual Offenses')
plt.legend(loc='lower right')
plt.show()
```

```python
# print slope, intercept, and coefficient of determination (r^2)
print('Estimated slope: {:.2f}'.format(linear_model.coef_[0]))
print('Estimated intercept: {:.2f}'.format(linear_model.intercept_))
print('Coefficient of Determination (r^2): {:.2f}'.format(linear_model.score(annual_stops.reshape(-1,1), annual_offenses)))
```

The estimated slope of this linear regression model is 0.14. This value indicates that as the number of annual stops increases by 1, the number of annual offenses is predicted to increase by 0.14.

We see here that number of stops is a moderate predictor of number of offenses overall. 
The r<sup>2</sup> value (the coefficient of determination) is 0.55, indicating that only 55% of the variation in the number of total annual offenses can be predicted by the number of annual stops.

#### **Defining Functions**

##### (a) Bootstrapping Function

```python
def bootstrap(df, colname):
    """ Calculate 5000 bootstraps on the colname column in df for a given year. """
    data= df[colname]
    sample_size= data.size
    
    # Bootstrap 5000 times. For each resample, calculate the mean of the sample.
    bootstraps= []

    for i in np.arange(0, 5001):
        resample= np.random.choice(data, sample_size)
        resample_mean= np.average(resample)
        bootstraps= np.append(bootstraps, resample_mean)
        
    return bootstraps
```

##### (b) Contingency Table Function

```python
def get_contingency_table(df, col1_name, col2_name):
    """ Generate a contingency table using the DataFrame df, and the 2 column names provided (Strings) col1_name and col2_name. """
    
    col_a_categories= df[col1_name].unique()
    col_a= pd.Categorical(df[col1_name], categories= col_a_categories)
    
    col_b_categories= df[col2_name].unique()
    col_b= pd.Categorical(df[col2_name], categories= col_b_categories)
    
    return pd.crosstab(col_a, col_b, rownames= [col1_name], colnames= [col2_name])
```

##### (c) Calculate Chi-Squared P-Value Function

```python
def calc_chi_squared(df, col1_name, col2_name):
    """ Calculate the chi-squared p-value of 2 columns from a given DataFrame. """
    cTable= get_contingency_table(df, col1_name, col2_name)
    
    return stats.chi2_contingency(cTable)[1]
```

##### d) Logistic Regression Prediction Accuracy Function

```python
def LogR(logdf, getbinary, targetvar):
    """ creates a df with only vars of interest with indicator variables, runs a logistic regression on that and returns a prediction accuracy percentage.
    Parameters: logdf is a df with only the variables in interest for logistic regression. 
    targetvar is a string of the target variable for the logistic regression.
    getbinary is a dataframe with columns that need to be given indicator variables and will then be added on to the logistic_data df.
      """
    newgetbinary = pd.get_dummies(getbinary)               # creating new df for log regression
    newlogdf = pd.concat([logdf, newgetbinary], axis=1)

    predictors = list(newlogdf.columns)
    predictors.remove(targetvar)
    train,test = train_test_split(newlogdf, test_size=0.2, random_state=15)
    logmod= LogisticRegression(solver='lbfgs', max_iter=1000);
    logmod.fit(train[predictors], train[targetvar])  
    tepreds=logmod.predict(test[predictors])
    
    return "the prediction accuracy as a percentage is {:.1f}.".format(np.sum(tepreds==test[targetvar])/len(tepreds)*100)
```

##### e) Predictors Coefficients Function

```python
def coeffs(logdf, getbinary, targetvar): 
    """Prints a list of coefficients and displays a plot of the coefficents in descending order.
    Parameters: logdf is a dataframe with only variables of intereset for logistic regression.
    getbinary is a dataframe with variables that need to be given indicator variables for running a  logistic regression.
    targetvar is a string of the target variable name for the logistic regression.
    """
    newgetbinary = pd.get_dummies(getbinary)
    newlogdf = pd.concat([logdf, newgetbinary], axis=1)  
    predictors = list(newlogdf.columns);
    predictors.remove(targetvar);
    train,test = train_test_split(newlogdf, test_size=0.2, random_state=15);
    logmod = LogisticRegression(solver='lbfgs', max_iter=1000);
    logmod.fit(train[predictors], train[targetvar]);
    
    # print the coefficients legibly
    for i, predictor in enumerate(predictors):
        print(f'{logmod.coef_[0,i]:.3f}\t{predictor}')

    # plot predictors by lean
    df = pd.DataFrame({'coef':logmod.coef_[0]}, index=predictors)
    sorted=df.sort_values(by='coef', ascending=False)
    sns.barplot(x=sorted['coef'],y=sorted.index, hue_order=sorted )
```

## VI. Interpretations and Conclusions

#### Overall Crime Trends

The linear regression analysis did not show a strong relationship between the number of stops made by NYPD officers per year and the total number of criminal offenses reported. The r<sup>2</sup> value (the coefficient of determination) of the model was 0.55, indicating that only 55% of the variation in the number of total annual offenses could be predicted by the number of annual stops. However, the NYPD's Stop and Frisk program primarily targets civilians suspected of drug- and/or weapons-related offenses. The second linear regression model, which focused specifically on drug and weapon crimes, demonstrated a much stronger relationship. The r<sup>2</sup> value of this model was 0.70, indicating that 70% of the variation in the number of annual drugs & weapons offenses can be predicted by the number of annual stops. 

The simple plot of the number of annual stops vs. annual drug & weapons offenses illustrated that this kind of illegal activity decreased dramatically over the course of the Stop and Frisk Program. However, the linear regression model confirmed that a significant fraction of of this observable change was a direct result of the NYPD's program.

#### Stop, Question, and Frisk Significance Tests

**Insignificant Predictor Variables**<br>
All significance tests (chi-squared tests and p-values generated by bootstraps) except for Frisked 2011 showed that observation duration and age had no significant dependence relationship with a person being frisked, searched, or arrested through the Stop, Question, and Frisk program. There are various reasons why this could be. Looking again at the scatterplots of observation duration in section IV.) Data Analysis shows that observation duration contained little variability across the entire sample. This trend could partially explain why observation duration is not a significant variable associated with whether one is frisked, searched, or arrested. The association between observation duration and all our target variables (frisked, searched, arrest made) may be insignificant because the times do not vary by much; the ranges are less than a minute for each target variable. The chi-squared tests determined observation duration has no association with whether one is frisked, searched, or arrested. This means there is no dependence relationship between observation duration and our target variables. Observation durations are roughly the same among all those who were and were not frisked, searched, or arrested through the Stop, Question and Frisk program in both 2011 and 2019.

Age was also an insignificant predictor for all target variables except for Frisked 2011. This was an interesting finding because it is contrary to what popular literature reports on how age interfaces with the Stop, Question, and Frisk Program. Literature surrounding the Stop, Question, and Frisk Program indicates that age could be a significant predictor for whether one is frisked, searched, or arrested. A [survey conducted by the Vera Institute of Justice](https://www.vera.org/downloads/Publications/coming-of-age-with-stop-and-frisk-experiences-self-perceptions-and-public-safety-implications/legacy_downloads/stop-and-frisk-technical-report-v4.pdf) found that 44% of young people surveyed in New York City reported being stopped by police repeatedly, meaning at least 9 times. There could be various reasons why we found age to be a generally insignificant predictor. One limitation of our project is that we do not possess the computing power to work with the full Stop, Question, and Frisk dataset for a given year. Instead, we sampled the datasets to generate a smaller subset of the population. Our sample could be less representative of the population just due to chance. Another possibility is the method in which we used to determine if age was a significant predictor of search made or arrest made. Due to the nature of the age data, we had to bootstrap the data and generate a p-value from the bootstrapped age means. Bootstrapping has its own limitations, including that it is best used for data that does not follow a normal distribution. There are also limitations to using mean age as the test statistic, as it is more easily skewed by outliers than a different test statistic such as the median age. Referring back to section IV.) Data Analysis, plotting the ages of the sample shows that age is right skewed. This skew of the age data means younger people were stopped more often than older people in both years. Since the age data was best analyzed using the bootstrapping method given the statistical knowledge of our team, there may be some limitations of our evaluation techniques that are unable to be mitigated in this project.

**Significant Frisked Predictor Variables**<br>
In 2011, the significance tests conducted showed that all 5 predictor variables had a significant dependence relationship with whether a person stopped was frisked. This means that race, sex, boro, age, and observation duration played a statistically significant role in whether or not a person stopped was frisked for weapons and/or drugs in 2011.

In 2019, the significance tests conducted showed that only the sex, race, and borough predictor variables had a significant dependence relationship with whether a person stopped was frisked. This means that sex, race, and borough played a statistically significant role in whether or not a person stopped was frisked for weapons and/or drugs in 2019.

**Significant Searched Predictor Variables**<br>
In 2011, the significance tests conducted showed that only the sex, race, and borough predictor variables had a significant dependence relationship with whether a person stopped was searched. This means that sex, race, and borough played a statistically significant role in whether or not a person stopped was searched in 2011.

In 2019, the significance tests conducted showed that only the sex predictor variable had a significant dependence relationship with whether a person stopped was searched. This means that sex played a statistically significant role in whether or not a person stopped was searched in 2019. This is an interesting finding because it is uncommon to find only one predictor variable with a significant relationship with a target variable.

**Significant Arrest Made Predictor Variables**<br>

## VII. Limitations

The two groups of data we have are datasets of the NYPD's Stop and Frisk instances for 2011 and 2019, and general crime data including major and minor felonies, misdemeanors, and violations in NYC for the years 2009-2019. Our stop and frisk data is only for NYC so we cannot use it as a representation of stop and frisks in other locations. Since the Stop and Frisk data is collected by the police officers who make the stops, the data is dependent on police officers’ judgement and possible bias. Officers make stops when they observe others and have suspicion to frisk them or probable cause to search them. The data is therefore limited in part to the individual officer’s judgment. For example, one officer may find a pedestrian suspicious and frisk them while another officer may not make that same judgement. In this way, there may be gaps or inconsistencies in our data in what one considers to be suspicious activity because it varies among officers and the accepted conditions of suspicious activities may also change throughout the years. The stop data is also limited by whether or not the officer decides to report it. There may be more stops made than we actually have in the data and we will not be able to analyze those unrecorded stops. Another limitation of this data is the missing data in many of the instances of the datasets. During our data cleaning section, we had to drop many instances where the age, observation duration, or borough information was poorly recorded or empty entirely. Those dropped instances do not have a chance to be considered in our analysis. While we cannot come up with a definitive answer to whether or not those dropped instances will make a significant difference to our analysis, we still must recognize their lost potential to affect our analysis.

The case for general crime data is slightly different when it comes to underreporting. Underreporting in a stop and frisk case is dependent on the officer while for general crime statistics, regular civilians may be involved in crime and choose not to report it. Reasons for the public underreporting crime can be due to a distrust in the police, whether one is an undocumented immigrant, or a prior settlement. According to an article written by the West Virginia University,  up to or above 90% or rape cases go unreported. The most common reason for not reporting rape are fear of not being believed, insecurity, and fear of getting into trouble.(1) We can conclude our data does not include all crime occurrences since it is dependent on civilians reporting or police officers reporting crime.
